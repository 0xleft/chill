```python
# load adata
import numpy as np
import os
from PIL import Image
from sklearn.model_selection import train_test_split
```

```python
# load data
data = []
labels = []

for file in os.listdir("data"):
    if file.endswith(".jpg"):
        try:
            img = Image.open("data/" + file)
            img = img.resize((150, 150))
            data.append(np.array(img))

            # open file that ends with .txt instead
            file = file.replace(".jpg", ".txt")
            with open("data/" + file) as f:
                label = f.readlines()
                # divide first by 90
                # divide second by 180
                label = [float(x) for x in label]
                labels.append(label)
        except:
            pass
```

```python
# transform data
import requests

new_labels = []
url = "https://nominatim.openstreetmap.org/reverse?lat=LATITUDE&lon=LONGITUDE&format=json"

print(len(labels))
# split to 8 threads
labels = np.array_split(labels, 30)

print(len(labels))

def get_country(in_labels, counter):
    for i in range(len(in_labels)):
        try:
            lat = in_labels[i][0]
            lon = in_labels[i][1]
            new_url = url.replace("LATITUDE", str(lat)).replace("LONGITUDE", str(lon))
            response = requests.get(new_url)
            print(response.json()["address"]["country_code"])
            print(i)
            new_labels.append({
                "country": response.json()["address"]["country_code"],
                "i": counter
            })
            counter += 1
        except:
            continue
```

```python
print(len(labels[9]))
```

```python
import threading
threads = []
total = 0
for i in range(30):
    t = threading.Thread(target=get_country, args=(labels[i], total, ))
    threads.append(t)
    t.start()
    total += len(labels[i])
    print(total)

for t in threads:
    t.join()
```

```python
# save to new_data
for i in range(len(new_labels)):
    np.save("new_data_2/" + str(i), data[new_labels[i]["i"]])
    with open("new_data_2/" + str(i) + ".txt", "w") as f:
        f.write(new_labels[i]["country"])
```
