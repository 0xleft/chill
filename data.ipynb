```python
# load adata
import numpy as np
import os
from PIL import Image
from sklearn.model_selection import train_test_split
```

```python
# load data
data = []
labels = []

for file in os.listdir("data"):
    if file.endswith(".jpg"):
        try:
            img = Image.open("data/" + file)
            img = img.resize((150, 150))
            data.append(np.array(img))

            # open file that ends with .txt instead
            file = file.replace(".jpg", ".txt")
            with open("data/" + file) as f:
                label = f.readlines()
                # divide first by 90
                # divide second by 180
                label = [float(x) for x in label]
                labels.append(label)
        except:
            pass
```

```python
# transform data
import requests

new_labels = []
url = "https://nominatim.openstreetmap.org/reverse?lat=LATITUDE&lon=LONGITUDE&format=json"

for i in range(len(labels)):
    lat = labels[i][0]
    lon = labels[i][1]
    new_url = url.replace("LATITUDE", str(lat)).replace("LONGITUDE", str(lon))
    response = requests.get(new_url)
    print(response.json()["address"]["country_code"])
    new_labels.append(response.json()["address"]["country_code"])
```

```python
# print labels
print(set(new_labels))
print(len(set(new_labels)))
```

```python
# save to new_data
for i in range(len(labels)):
    np.save("new_data/" + str(i), data[i])
    with open("new_data/" + str(i) + ".txt", "w") as f:
        f.write(new_labels[i])
```
